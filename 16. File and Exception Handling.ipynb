## Exception Handling
* Exception disrupts the flow of the program and terminates the program abruptly.
* Exception needs to be handledm using exception handling.
* We want to handle this exception using try... except... block.

```python
# ZeroDivisionError
a = 5
b = 0
try:
    c = a/b
    print(c)
except ZeroDivisionError as e:
    print(e)
```

```python
# When exception name is unknown
a = 5
b = 0
try:
    c = a/b
    print(c)
except Exception as e:
    print(type(e)) # prints type of error
    print(e)
```

## Raising an exception
* It is used to inform the user before the program starts executing.

```python
# Raising error
a = 5
b = 0
# b = 2

if b == 0:
    raise ValueError("b cannot be zero")
else:
    c = a/b
    print(c)
```

## File Handling

## open(filename, mode)
* Types of mode
  * create : "x"
  * read : "r"
  * write : "w"
  * append : "a"

## create

```python
# create file
# variableName = open(filename, create mode)

file = open("ast.txt","x") # step:1
print("file created", file)
file.close()                  # step:2

# check the folder on side bar, file is created there
```

## read

```python
import os # this library is used to check whats in the directory
os.path.exists("fileName.txt")
```

```python
# Helps list all the files present in the directory
os.listdir()
```

```python
# reading file "test.txt" that doesnot exist in the directory, throws error
file = open("test.txt","r")

file.close()
```

## write
* In write mode if the file doesnot exist, then new file is created.
* If the file already exists, then old file is overwritten. The old data in the file is infact lost, or overwritten with new one.

```python
# new file created as no such file exists
file = open("city.txt", "w")
file.write("Patan")
file.close()
```

```python
# reading the file
file = open("city.txt", "r")
text = file.read()
file.close()
print(text)
```

```python
# old file with same name exists, so data is overwritten with new data
file = open("city.txt", "w")
file.write("Bhaktapur")
file.close()
```

## append
* In append mode if the file doesnot exist, then new file is created.
* If the file already exists, then old file is not overwritten, unlike write mode. The old data in the file is intact, or not overwritten with new data. Rather new data is added/appended at the end.

```python
# new file created as no such file exists
file = open("city2.txt", "a")
file.write("Patan \n")
file.close()
```

```python
# old file with same name exists, so new data is appended or concatenated
file = open("city2.txt", "a")
file.write("Bhaktapur \n")
file.close()
```

## with open(filename, mode)
* syntax:: with open(filename, mode) as variableName:
* Easier than "open()", as "with open()" automatically closes the files. No need to close each time after operation.

## read file

```python

with open("city2.txt", "r") as file:
    text = file.read()

print(text)
```

## If want output as a list using `readlines()`

```python
with open("city2.txt", "r") as file:
    text = file.readlines()

print(text)
```

## If want to read only line using `readline()`

```python
with open("city2.txt", "r") as file:
    text = file.readline()

print(text)
```
