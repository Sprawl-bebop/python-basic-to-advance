## STRING IS IMMUTABLE
|h|e|y|
|---|---|---|
| [1004] |

### |~~h~~ k | e | y | [1008]


- If want to `update` or `edit` the given string, then it is not allowed as string is `immutable`
- `immutable` suggests that it is not possible to edit or update the string in the same memory address.
- Rather it is copied to another memory locataion and variable is updated with newe memory address.

## String after updation is copied to new memory location. And the variable points to the new memory address. String doesnot refer/point (disassociation with old memory address) to the old memory location. This is the concept of `immutable`

```python
str = "hey"
print(id(str))

#after updating
str = "key"
print(id(str))
```

## 1. `replace()` method
- This is used in `data cleaning`.
- User may use some special character instead of alphabets. In such case those characters need to be replaced. `@` instead of `a`, `$` instead of `s`
- This method relies on the concept of `immutable`

```python
# we need to replace @ with a
name = "@shrit"
print(name, id(name))

# strings are immutable, hence after update, variable points to new memory address
name = name.replace("@", "a")
print(name, id(name))
```

```python
# replacing all the character
name = "@t@dy"
print(name, id(name))

name = name.replace("@", "a")
print(name, id(name))
```

```python
# method chaining
name = "@$hrit"
print(name, id(name))

# here using multiple methods
# output (a$hrit) of first method is passed as an input to another method which now replaces "$" with "s"
name = name.replace("@", "a").replace("$", "s")
print(name, id(name))
```

## 2. `upper()` `lower()` `title()` methods
- Very important for ML
- `upper()` is used to change case to uppercase
- `lower()` is used to change case to lowercase
- `title()` is used to capitalize the starting letter of each word

```python
# .upper() method
text = "hello world"

# without updating variable
print(text.upper())
print(id(text))

# updating variable
text = text.upper()
print(text)
print(id(text))
```

```python
# .lower() method
text = "HeLLo WorlD"

# updating variable
text = text.lower()
print(text)
```

```python
# .title() method
text = "hello world"

text = text.title()
print(text)
```

## 3. `strip()` `lstrip()` `rstrip()` methods
- `strip()` is used to remove extra spaces from both extreme left and extreme right.
- `lstrip()` is used to remove extra spaces from only extreme left.
- `rstrip()` is used to remove extra spaces from only extreme right.


```python
# .strip() method
# the given string has extra spaces on both left and right
# using .strip() method removes those extra spaces on both left and right, note not from between words
msg = "    hello world     "
msg = msg.strip()
print(f'"{msg}"')
```

```python
# .lstrip() method
# the given string has extra spaces on both left and right
# using .lstrip() method removes those extra spaces on only from the left
msg = "    hello world     "
msg = msg.lstrip()
print(f'"{msg}"')
```

```python
# .rstrip() method
# the given string has extra spaces on both left and right
# using .rstrip() method removes those extra spaces on only from the right
msg = "    hello world     "
msg = msg.rstrip()
print(f'"{msg}"')
```

## 4. `split()` `join()` methods
- `split()` is used to split or extract words from the given string based on the `delimeter` defined. It could be any white space or special characters.
- `join()` is used to join the words from a given `List` to form a sentece based on the defined `delimeter`.
- Note: `type` of output of `split()` is a `LIST`

```python
# split() method
# spliting given string based on "space" as a "delimeter"
text = "hello world again"
words = text.split(" ")
print(words)
type(words)
```

```python
#  join() method
# join() is used for joining from given set of list using a "delimeter"
# join() allows only "list of string"
word_list = ["testing", "testing", "1", "2" , "3"]
msg = " ".join(word_list)
print(msg)
type(msg)
```

```python
# join() uses "delimeter" of any character
word_list = ["hello", "hello", "1", "2" , "3"]
msg = "_".join(word_list)
print(msg)
type(msg)
```
