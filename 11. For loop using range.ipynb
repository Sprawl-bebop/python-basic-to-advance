## Range function
- range(`start`, `stop`, `step`)
  - `start` default value is 0
  - `step` default value is 1
  - `stop` value is not included, only value below stop is taken

```python
#range(start, stop, step)
for index in range(0,5,1):
    print(index)
```

## Range default values and short hand
- `range(5)`
- Here, `start` starts at default 0, `step` has default value as 1, `stop` is taken as 5

```python
for index in range(5):
    print(index)
```

```python
# start=2, stop=10 ie value below 10, step=2
for index in range(2,10,2):
    print(index)
```

## `string` and `for` loop using range

```python
# accessing element of certain index
s = "hello"
index = 1
letter = s[index]
print(f"the index is {index} and letter is {letter}")
```

```python
# using for to loop through string
s = "hello"

for index in range(len(s)):
    letter = s[index]
    print(f"the index is {index} and letter is {letter}")
```

## WAP to check if a letter is `vowel` or `consonant`

```python
s = "hello"

for index in range(len(s)):
    if s[index] == "a" or s[index] == "e" or s[index] == "i" or s[index] == "o" or s[index] == "u":
        print(f"the letter {s[index]} is vowel")
    else:
        print(f"the letter {s[index]} is consonant")
```

## `continue` and `break` KEYWORDS
- in a `for` loop, `continue` disrupts the loop rythmn. It stops and skips current iteration and doesnot execute the code below it. Note, it continues with the remaining loop.
- in a `for` loop, `break` stops the loop. It stops and exits after `break` keyword is detected and also doesnot execute the code further below it. Note, it breaks the remaining loop promptly and doesnot execute the loop further.

```python
# continue example
for index in range(5):
    if index == 2:
        print("Condition is met, code excutes up to here. Continue breaks this iteration and doesnot print 2. But for loop continues.")
        continue
        print("this doesnot get executed")
    print(index)

print("\nBlock outside of loop")
```

```python
# break example
for index in range(5):
    if index == 3:
        print("condition is met. After break, code below is not excuted. Also for loop is terminated!!")
        break
    print(index)

print("\nBlock outside of loop")
```

## WAP to check vowel or consonant. Here white space is detected and is ignored using `continue` keyword.

```python
# continue
s = "hello world"

for index in range(len(s)):
    if s[index] == "a" or s[index] == "e" or s[index] == "i" or s[index] == "o" or s[index] == "u":
        print(f"the letter {s[index]} is vowel")
    elif s[index] == " ":
        continue
    else:
        print(f"the letter {s[index]} is consonant")
```

## WAP to count `vowel` or `consonants` on the given string `[Assignment]`

```python
str = input("Enter string\n")
str = str.lower()

str_vowel = []
str_cons = []

vowel_count = 0
consonant_count = 0

for index in range(len(str)):
    if str[index] == "a" or str[index] == "e" or str[index] == "i" or str[index] == "o" or str[index] == "u":
        vowel_count += 1
        str_vowel.append(str[index])
    elif str[index] == " ":
        continue
    else:
        consonant_count += 1
        str_cons.append(str[index])

print(f"Given string: '{str}'\nVowel Count: {vowel_count} \t Consonant Count:{consonant_count}")
print("Vowels:",str_vowel)
print("Consonants:",str_cons)
```
