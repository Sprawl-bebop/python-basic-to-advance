## Using if only. each if is checked individually

```python
n = 45

if n < 0:
    print("negative")
if n == 0:
    print("neutral")
if n > 0:
    print("positive")
```

## Nesting

```python
n = 0

if n < 0:
    print("negative")
else:
    if n == 0:
        print("neutral")
    else:
        print("positive")
```

## Chaining using else if (elif), most preferred

```python
n = 0

if n < 0:
    print("negative")
elif n == 0:
    print("neutral")
else:
    print("positive")
```
