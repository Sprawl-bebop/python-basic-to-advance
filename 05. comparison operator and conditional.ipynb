## WAP to check if input number is negative or positive

```python
num = int(input("Enter a number\n"))
if num >=0:
  print(f"{num} is a positive number.")
else:
  print(f"{num} is a negative number.")
```

## WAP to display "Hello" if a number entered by user is a multiple of 5, otherwise print "Bye".

```python
num = int(input("Enter a number\n"))
if num%5 == 0:
    print("\nHello")
else:
    print("\nBye")
```

## WAP to check if the user given input is odd or even. If it is even print even else print odd.

```python
num = int(input("Enter a number\n"))
if num%2 == 0:
    print(f"Given number {num} is even.")
else:
    print(f"Given number {num} is odd.")
```

## WAP to check whether the last digit of a number entered by user is divisible by 3 or not.

```python
num = int(input("Enter a number\n"))

temp = num%10

if temp%3 == 0:
    print(f"Last digit of entered number {num} is {temp}.\nIt is divisible by 3.")
else:
    print(f"Last digit of entered number {num} is {temp}.\nIt is not divisible by 3.")
```
