## WAP to check the age category

```python
age = int(input("Enter age\n"))

if age < 13:
    print("you are a child")

elif age >= 13 and age <= 19:
    print("you are a teen")

elif age >= 20 and age <= 50:
    print("you are an adult")

else:
    print("you are an elderly")
```

## WAP to check score bracket

```python
score = int(input("Enter your score\n"))

if score == 80 or score > 80:
    print("you got distinction")

elif score >= 60 and score < 80:
    print("you got first division")

elif score >= 40 and score < 60:
    print("you got second division")

else:
    print("Sorry, you failed")
```

## Body Mass Index(BMI) calculator, weight in kg, height in cm.
- $BMI = \frac{weight}{height^2}$

<img src="https://nutrition.health.gov.lk/wp-content/uploads/2020/12/BMI-1024x569.png" height="250">

```python
import math

weight = float(input("Enter your weight(kg)\n"))
height = float(input("Enter your height(cm)\n"))

height = height/100

bmi = round(float(weight / math.pow(height,2)),3)

print(f"\nYour Body Mass Index(BMI) is {bmi}\n")

if bmi < 18.5:
    print("You are underweight")

elif bmi >= 18.5 and bmi <= 22.9:
    print("You are normal")

elif bmi >= 23 and bmi <= 24.9:
    print("You are at risk to overweight")

elif bmi >= 25 and bmi <= 29.9:
    print("You are overweight")

elif bmi >= 30:
    print("You are suffering from obesity")
```
