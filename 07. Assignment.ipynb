## 1. WAP to swap values between two variables
* num1 = 2.5
* num2 = 0.36

```python
num1 = 0.25
num2 = 0.36
print(f"Before swap:\n num1: {num1} \t num2: {num2}")
num1 += num2
num2 = num1 - num2
num1 -= num2
print(f"\nAfter swap:\n num1: {num1} \t num2: {num2}")
```

```python
num1 = 0.25
num2 = 0.36
print(f"Before swap:\n num1: {num1} \t num2: {num2}")
temp = num1
num1 = num2
num2 = temp
print(f"\nAfter swap:\n num1: {num1} \t num2: {num2}")
```

```python
num1 = 0.25
num2 = 0.36
print(f"Before swap:\n num1: {num1} \t num2: {num2}")
num1, num2 = num2, num1
print(f"\nAfter swap:\n num1: {num1} \t num2: {num2}")
```

## 2. The volume of sphere with radius r is $\frac{4}{3}\pi r^3$. What is the volume of sphere with radius r taken from user input.

```python
import math
r = float(input("Enter radius of a circle:\n"))
volume = float((4/3)*math.pi*r**3)
print(f"Volume of sphere with radius {r} is {round(volume,3)}.")
```

## 3. How many seconds are there in 42 minutes 42 seconds? Or take user input for minutes and seconds.

```python
input_minute = int(input("Enter time in minutes\n"))
input_second = int(input("Enter time in seconds\n"))
time_in_seconds = input_minute * 60 + input_second
print(f"\nTotal Time in seconds: {time_in_seconds}")
```

## 4. How many miles are there in 10 kilometers or take input from user? Hint: there are 1.609 km in a mile.

```python
distance_km = float(input("Enter distance in kilometers\n"))
distance_mile = float(distance_km / 1.609)
print(f"{distance_km} km = {round(distance_mile, 3)} mile")
```

## 5. If you run a 10 kilometer race 42 minute 42 second, what is your average speed.
[speed: mile in minute and mile in seconds]

```python
distance_km = 10
minute = 42
second = 42

#for mile in minute
distance_mile = float(distance_km / 1.609)
time_in_minute = float(minute + second/60)
speed_mile_minute = float(distance_mile / time_in_minute)
print(f"mile:{distance_mile} minute:{time_in_minute} \nAverage Speed: {round(speed_mile_minute, 3)} mile/min\n")

#for mile in second
distance_mile = float(distance_km / 1.609)
time_in_second = float(minute*60 + second)
speed_mile_second = float(distance_mile / time_in_second)
print(f"mile:{distance_mile} second:{time_in_second} \nAverage Speed: {round(speed_mile_second, 4)} mile/sec")
```
