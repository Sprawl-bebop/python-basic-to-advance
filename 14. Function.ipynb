## FUNCTIONS
``` python
 def function_name(argument_1,argument_2,argument_3,....,argument_n):
    # function body
    return expression
```

````python
def addition():
    a = 5
    b = 10
    total = a + b
    print(total)

# calling function, which has no return type, and just prints on console
addition()

# there is no return so it will be  assigned with none
output = addition()
print(output)
````

````python
def add():
    a = 5
    b = 10
    total = a + b
    return total

# with return, varible output is assigned with the return value from function
output = add()
print(output)
````

````python
#passing argument
def addition(a, b):
    total = a + b
    return total

# not best practice
output = addition(5,5)

# best practice
output = addition(a=5,b=5)

print(output)
````

## WAP to check if the given number is prime or not using a function.

````python
def is_Prime(num):
    for i in range(2, round(num/2)):
        if num % i == 0:
            return False
      # why else is not written, IMPORTANT. As else will return false,
      # which is not the "False" we require for prime
      # else:
      #    return True

    return True

is_Prime(13)
````

## Default vs Non-default
- Non-default is when there is no value assigned to the parameters in the function parameter section. Example: greet(name, msg):
- Default assigns the parameter with specific value. Reasons could be the value is kept default in every case and user dont need to provide with every input.
Example: greet(name, msg="Hi"):
- `NOTE` Conditions:
  - `IMP`All non-default is mentioned first, and then only followed by default declaration.
  - `IMP`After default argument, non-default is not allowed.
  - All can be non-default.
  - All can be default.

````python
# non-default
def greet(name, msg):
    return f"Hi {name}, {msg}"

greet("ashrit", "hello")
````

````python
# default
def greet(name, msg="Good day"):
    return f"Hi {name}, {msg}"

greet("ashrit")
greet("ram")
````

## Local vs Global

````python
# local variable
name = "ram"

def callMe():
    name = "ram2"
    print(name)

callMe()
````

````python
# bad practice, do not call global this way
name = "ram"

def callMe():
    # accessed the global ram, and reassigned it, this way is not a good practice
    global name
    name = "changed ram"
    print(name)

callMe()
````

## WAP to find a factorial using a function.

````python
def factorial(num):

    if num == 0 or num == 1:
        return 1

    total = 1
    for i in range(2,num+1):
        total *= i
    return total

factorial(num=6)
````

## `*args` and `**kwargs`
- `*args` and `**kwargs` are special keywords.

## `*args`
- `*args` refers to arguments.
- `*args` is similar to over-riding but simplified. It allows a function to accept multiple parameters from a user through the same function, extending its reusability and flexibility.
- `*args` accepts parameters as a `tuple`.

````python
# general example
def test(*args):
    print(args) # note: only args is written, * is omitted
    print(type(args))

test(1, 2, 3, 78)
````

````python
# More example

def add(*args):
    total = 0
    for item in args: # note: only args is written, * is omitted
        total = total + item

    return total

# same function can be used with multiple pararmeters, method over-riding
print(add(2,3))
print(add(2,3,67,89))
print(add(2,3,5,6,65,4,9,12))
````

## `**kwargs`
- `**kwargs` refers to key-word-arguments i.e. key-value-pair.
- `**kwargs` accepts parameters as a `dictionary`.
- `**kwargs` is similar to `*args`. It allows a function to accept multiple parameters but it must be in key-value pair.

````python
# general example
def test(**kwargs):
    print(kwargs) # note: only kwargs is written, ** is omitted
    print(type(kwargs))

test()
````

````python
# Error example
def test(**kwargs):
    print(kwargs) # note: only kwargs is written, ** is omitted
    print(type(kwargs))

test("name","age") # this must be key-pair
````

````python
# Correct Example
def test(**kwargs):
    print(kwargs) # note: only kwargs is written, ** is omitted
    print(type(kwargs))

test(name="ram",age=23) # this must be key-pair
test(name="ram",age=23, address="test") # with more parameters
````

## Multiple `return` values from the same function.

````python
# WAP to calculate gcd and lcm returning both values from the same function
import math
def calc(num1, num2):
    gcd = math.gcd(num1, num2)
    lcm = math.lcm(num1, num2)

    return  lcm, gcd

# calc(num1=12, num2=8)
print(calc(num1=12, num2=8)) # we get output as a tuple, but we want individual values

# Using tuple unpacking, we will assign individual values to variables

# note: order pf the return 1st:lcm 2nd:gcd
lcm, hcf = calc(num1=12, num2=8)
# hcf, lcm = calc(num1=12, num2=8) #error because of order of return, wrong values

print(f"HCF:{hcf} LCM:{lcm}")
````

## Recursion
- Calling itself.
- Must have a stopping condition suggesting when to stop avoiding infinite looping.

``` python
    # case of recursion with infinite looping, which doesnot have stopping condition
    def test():
        test()
```


````python
# WAP to find a factorial using recursion
def factorial(n):
    if n == 1 or n == 0:
        return 1
    else:
        return n * factorial(n-1)

factorial(n=5)
````

## WAP to find the fibonnaci number of the given term using recursion.

````python
def fib(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)

fib(n=5)
````
